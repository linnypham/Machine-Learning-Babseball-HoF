## Predicting Hall-of-Fame players

WAR stat of all players

```python
import pandas as pd
from pybaseball import bwar_bat,bwar_pitch,playerid_lookup

war_bat = bwar_bat()
war_pitch = bwar_pitch()

war_bat.to_csv('data/bwar_bat.csv',index=False)
war_pitch.to_csv('data/bwar_pitch.csv',index=False)
```

HoF Player ID

```python
hof_player = pd.read_csv('data/hof_players.csv')
hof_name = hof_player['Player']
hof_data = []
for fullname in hof_name:
    name = fullname.split()
    data = playerid_lookup(name[1],name[0])
    hof_data.append(data)
print(hof_data)
'''hof_data.to_csv('data/hof_stats.csv')'''
```
